```python
import numpy as np
from scipy.stats import pearsonr
R = np.array([[1, 2, 0, 2, 5], 
           [2, 2, 5 ,5, 5],
           [1, 0, 5, 2, 4],
           [1, 5, 5, 2, 1]])
R
```

```python
unknown_product = []
for i in range(0, R.shape[0]):
    unknown_product.append(np.where(R[i] == 0))
unknown_product[0][0]
```

```python
for i in range(0, len(unknown_product)):
    if len(unknown_product[i][0]) > 0:
        for j in range(0, R.shape[0]):
            remove_index_array = np.unique(np.concatenate((unknown_product[i][0], unknown_product[j][0])))
            matrixA = np.delete(R[i], remove_index_array)
            matrixB = np.delete(R[j], remove_index_array)
            corr, _ = pearsonr(list(matrixA), list(matrixB))
            print("Sim(X{}, X{}) = {}".format(i+1, j+1, corr))
```
